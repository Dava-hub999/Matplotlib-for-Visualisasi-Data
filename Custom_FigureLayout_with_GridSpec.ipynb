# Pengaturan Figure Layout dengan gridspec

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib.gridspec as gridspec
```

#### Simple 2x2 layout dengan subplots

```python
siswa = ['cecep', 'joko', 'wati', 'ani']
uas = [68, 83, 96, 72]

fig, axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True) # untuk membentuk agar bisa menampung 2 colom dan 2 baris (constraine_layout digunakan untuk mengatur jarak per gambar agar tidak berdempetan)

axes[0][0].bar(siswa, uas)
axes[0][1].pie(uas, labels=siswa)
axes[1][0].pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
axes[1][1].barh(siswa, uas)

plt.show()
```

#### Simple 2x2 layout dengan Gricspec

```python
fig = plt.figure(constrained_layout=True) # objek figure( constraine_layout digunakan untuk mengatur jarak per gambar agar tidak berdempetan)

gs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig) # (membentuk layout 2 x 2, figure=fig : untuk menampung object figure yang baru dibentuk)

# membentuk Object Axes
ax1=fig.add_subplot(gs[0, 0]) # mengature posisi layout
ax2=fig.add_subplot(gs[0, 1])
ax3=fig.add_subplot(gs[1, 0])
ax4=fig.add_subplot(gs[1, 1])

# menentukan gambar plot yang akan digunakan
ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)

plt.show()
```

#### Gridspec Model ke-2 ( lebih efisien )

```python
fig = plt.figure(constrained_layout=True) # objek figure( constraine_layout digunakan untuk mengatur jarak per gambar agar tidak berdempetan)

# ini yang membedakan dari model ke-1
gs = fig.add_gridspec(ncols=2, nrows=2) # (membentuk layout 2 x 2)

# membentuk Object Axes
ax1=fig.add_subplot(gs[0, 0]) # mengature posisi layout
ax2=fig.add_subplot(gs[0, 1])
ax3=fig.add_subplot(gs[1, 0])
ax4=fig.add_subplot(gs[1, 1])

# menentukan gambar plot yang akan digunakan
ax1.bar(siswa, uas)
ax2.pie(uas, labels=siswa)
ax3.pie(uas, labels=siswa, explode=[0, 0, 0, 0.2])
ax4.barh(siswa, uas)


plt.show()
```

```python
# Example ke-2

fig = plt.figure(constrained_layout=True) # objek figure( constraine_layout digunakan untuk mengatur jarak per gambar agar tidak berdempetan)

# ini yang membedakan dari model ke-1
gs = fig.add_gridspec(ncols=3, nrows=3) # (membentuk layout 3 x 3)

# membentuk Object Axes
ax1 = fig.add_subplot(gs[0, :]) # mengature posisi layout
ax1.set_title('gs[0, :]')

ax2 = fig.add_subplot(gs[1,:-1])
ax2.set_title('gs [1, :-1]')
ax2.barh(siswa, uas)

ax3 = fig.add_subplot(gs[1:,-1])
ax3.set_title('gs [1:, -1]')
ax3.pie(siswa, uas)

ax4 = fig.add_subplot(gs[-1,0])
ax4.set_title('gs [-1, 0]')

ax5 = fig.add_subplot(gs[-1,-2])
ax5.set_title('gs [-1, -2]')

plt.show()
```
