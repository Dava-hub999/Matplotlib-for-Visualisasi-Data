```python
# import module
import matplotlib.pyplot as plt 
import numpy as np
```

### simple line plot

```python
x = np.arange(0.0, 2.0, 0.01) # (start, stop, interval(jarak))
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots() # 00 style

ax.plot(x, s) # untuk penanda sumbu semacam (x, y)
ax.set(xlabel = 'nilai x',
        ylabel = 'nilai sin(x)',
        title= 'Visualisasi nilai sin')

ax.grid() # untuk menunjukkan kotak2 nya itu

plt.show()
```

```python
fig, ax = plt.subplots() # pyplot style

plt.plot(x, s) # untuk penanda sumbu semacam (x, y)
plt.plot(x, s, color='red')

plt.xlabel ('nilai x') 
plt.ylabel ('nilai sin(x)')
plt.title  ('Visualisasi nilai sin')

plt.grid() # untuk menunjukkan kotak2 nya itu

plt.show()
```

## Multiple subplots

```python
x1 = np.linspace(0.0, 5.0, 100) # (start, stop, jumlah data yg ingin dihasilkan)
x2 = np.linspace(0.0, 2.0, 100) # (start, stop, jumlah data yg ingin dihasilkan)

y1 = np.cos (2* np.pi * x1)
y2 = np.cos (2* np.pi * x2)
```

```python
x1
```

```python
# OOS style                                                                                                                                                                                                                                                                                         
fig, (ax1, ax2) = plt.subplots(2,1) # ( Jumlah baris , Jumlah kolom)
fig.suptitle(' Multiple Subplots dengan OO Style ')

ax1.plot(x1, y1, 'ro-') 
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_xlabel('Nilai x')
```

```python
# Pyplot style
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'ro-') 
plt.title(' Multiple Subplots dengan OO Style ')
plt.ylabel('Nilai $cos(x_2)$')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'g.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai x')

plt.show()
```
