## Animasi Pada Matplotlip

```python
fig = plt.figure() # container besar
plt.xlim((0, 4)) # untuk mengatur jangkauan sumbu x
plt.ylim((-2, 2)) # untuk mengatur jangkauan sumbu y

line = plt.plot([])[0]

def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x,y)
    return line,

anim = FuncAnimation(fig, animate, frames=500, interval=20) #(frames : mengatur pola smooth / , interval : mengatur kecepatan)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)

plt.close()
```
