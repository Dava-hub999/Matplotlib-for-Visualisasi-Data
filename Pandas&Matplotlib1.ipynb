### Visualisasi data yang tersimpan pada pandas data frame(bagian 1)

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

#### Sample Data

```python
# Data Dictionary
df = pd.DataFrame({
    'nama' : ['cecep', 'agus', 'anin', 'naela', 'dapa'],
    'usia' : [20, 19, 23, 25, 21],
    'gender' : ['L', 'L', 'P', 'P', 'L'],
    'propinsi' : ['jatim', 'jabar', 'jateng', 'lampung', 'bali'],
    'n_anak' : [2,3,1,5,4],
    'n_pets' : [8,5,6,3,9]
})

df
```

#### Scatter PLot

```python
df.plot(kind='scatter', x='n_anak', y='n_pets', color='red') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('Scatter Plot')
plt.xlabel('Jumlah anak') 
plt.ylabel('Jumlah peliharaan')

plt.show()
```

### Histogram

```python
df[['usia']].plot(kind = 'hist', bins = [0,20, 40, 60, 80, 100])

plt.title('Histogran ')
plt.xlabel('Usia') 
plt.ylabel('Jumlah')

plt.show()
```

### Line Plot

```python
df.plot(kind='line', x='nama', y='n_anak', color='green') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('Line Plot')
plt.xlabel('Nama ') 
plt.ylabel('Jumlah anak')

plt.show()
```

# Multiple LinePlot

```python
# Sebelum Multiple 

df.plot(kind='line', x='n_anak', y='n_pets', color='red') # (menentukan jenis plotting, data X, data Y, warna)
df.plot(kind='line', x='nama', y='n_anak', color='green') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('Scatter Plot')
#plt.xlabel('Jumlah anak') 
plt.ylabel('Jumlah peliharaan')

plt.show()
```

```python
# Sesudah

ax = plt.gca() # method agar bisa 1 axes

df.plot(kind='line', x='n_anak', y='n_pets', color='red', ax=ax) # (menentukan jenis plotting, data X, data Y, warna, menyisipka ax)
df.plot(kind='line', x='nama', y='n_anak', color='green', ax=ax) # (menentukan jenis plotting, data X, data Y, warna)

plt.title('Scatter Plot')
#plt.xlabel('Jumlah anak') 
plt.ylabel('Jumlah peliharaan')

plt.show()
```

### Bar Plot

```python
df.plot(kind='bar', x='nama', y='usia', color='red') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('Bar Plot Plot')
# plt.xlabel('Jumlah anak') 
plt.ylabel('Usia')

plt.show()
```

```python
# Pengelompokan berdasarkan kolom propinsi
# menghitung jumlah nama untuk setiap kelompok per propinsi
# lalu di plotting
df.groupby('propinsi')['nama'].count().plot(kind='bar')

plt.title('Bar Plot')
plt.xlabel('nama propinsi') 
plt.ylabel('banyak nama')

plt.show()
```

#### Stacked bar plot dengan 2 level group by

```python
# Pengelompokan berdasarkan kolom propinsi dan gender
# mengatur 
# lalu di plotting
df.groupby(['propinsi','gender']).size().unstack().plot(kind='bar', stacked=True)

plt.title('Stacked Bar Plot')
plt.xlabel('nama propinsi') 
plt.ylabel('banyak nama')

plt.show()
```

```python
# Pengelompokan berdasarkan kolom propinsi dan gender
# mengatur 
# lalu di plotting
df.groupby(['gender','propinsi']).size().unstack().plot(kind='bar', stacked=True) # (mengatur jenis, mengatur tumpuk atau tidak)

plt.title('Stacked Bar Plot')
#plt.xlabel('nama propinsi') 
plt.ylabel('Jumlah')

plt.show()
```
