# Box Plot / Box Chart


```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Sample Dataset

```python
np.random.seed(2) # fungsi ini digunakan untuk menghasilkan angka acak yang sama setiap kali Anda menjalankan kode yang menggunakan angka acak dari pustaka NumPy sedangkan seed(2) untuk nilai awalan saja tapi saya juga ga faham tapi intinya mah si method random tidak akan berubah ubah setiap kali di run ulang

data = np.random.normal(loc=100, scale=10, size=200)
# setelah menjalankan kode ini, variabel data akan berisi array dengan 200 (jumlah data yang dihasilkan) angka acak yang diambil dari distribusi normal dengan rata-rata 100 dan deviasi standar 10. 

data
```

### Simple Box plot / sederhana
##### boxplot 

```python
plt.boxplot(data, labels=['data']) # (manggil data, labels ini opsional (untuk mengatur penanda sumbu x))

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

### Notched Boxplot
#####  varian dari box plot yang mencakup elemen "notch" untuk memberikan informasi tambahan tentang keyakinan interval median(nilai tengah). Hal ini bisa sangat bermanfaat dalam perbandingan data antara dua kelompok atau lebih.

```python
plt.boxplot(data, labels=['data'], notch=True) # (manggil data, labels ini opsional (untuk mengatur penanda sumbu x), untuk manggil notch)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

#### Pengaturan marker untuk outlier(data yang diluar jangkauan 1,5 dari nilai IQR) 

```python
plt.boxplot(data, 
            labels=['data'], 
            showfliers=True, # untuk mengatur outlier dimunculkan atau tidak
            flierprops={'markerfacecolor':'r', 'marker' : 'd'} # harus membuat dict{untuk pengaturan warna, untuk pengaturan bentuk}
) 

plt.title('pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

### Horizontal boxplot

```python
plt.boxplot(data, labels=['data'], vert=False) # (manggil data, labels ini opsional (untuk mengatur penanda sumbu x), mengatur jenis atau hal hal lain pada boxplot)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

### Multiple Box plot

```python
# Multiple
np.random.seed(10)
data1 = np.random.normal(100, 10, 200) # (nilai rata-rata (mean) dari distribusi normal Artinya: data yang dihasilkan akan memiliki rata-rata 100, sebaran deviasinya yang berarti sebagian besar data akan berada dalam kisaran sekitar 90 hingga 110 (rata-rata  1 deviasi standar), Jumlah datanya)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4] #list untuk penampung keseluruhan 
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4'] # list untuk nilai( untuk memberi label di gambar)
```

```python
plt.boxplot(data, labels=labels) # (manggil data, labels ini opsional (untuk mengatur penanda sumbu x / ditampilkan di grafik))

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```
