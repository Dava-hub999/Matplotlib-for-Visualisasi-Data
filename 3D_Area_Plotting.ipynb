# 3D Area Plotting

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# datset sample
n = np.linspace(-3.0, 3.0, 100) # (nilai terkecil 0, nilai terbesar 1, banyak data )

x,y = np.meshgrid(n, n) # menghasilkan data 2 dimensi (data1 jadi x dan data2 jadi y)
z = np.sqrt(x**2 + y**2) # (100 * 100) dan menampung nilai quadratnya

print(x.shape)
print(y.shape)
print(z.shape)
```

```python
fig = plt.figure() # OO Style
ax = plt.axes(projection = '3d') # menampung fungsi 3D

ax.plot_wireframe( x,y,z, color='green') # untuk visualisasi data


ax.set( # untuk setting seperti biasanya
    title = ' Example 3D Scatter Plot ',
    xlabel = 'Sumbu X',
    ylabel = 'Sumbu Y ',
    zlabel = 'Sumbu Z'
)

plt.show()
```

# Surgaee plis

```python
fig = plt.figure() # OO Style
ax = plt.axes(projection = '3d') # menampung fungsi 3D

ax.plot_surface( x,y,z, color='red') # pembentukan line plot
ax.plot_surface( x,y,z, cmap='viridis') # pembentukan line plot

ax.set( # untuk setting seperti biasanya
    title = ' Example 3D Line Plot ',
    xlabel = 'Sumbu X',
    ylabel = 'Sumbu Y ',
    zlabel = 'Sumbu Z'
)

plt.show()
```
