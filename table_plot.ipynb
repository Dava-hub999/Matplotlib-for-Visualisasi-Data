# Table plot atau Table Chart

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Sample Data

```python
data_uas = [
    ['bejo', 70],
    ['tejo', 83],
    ['cecep', 62],
    ['zaenab', 74],
    ['naela', 71]
]
data_uas
```

### Table plot OO Style

```python
fig, ax = plt.subplots() #untuk menghasilkan 2 object yang di tampung juga dengan 2 variable

table = plt.table(
    cellText = data_uas, # 
    loc='center' # untuk mengatur lokasi dari table plot
)

table.set_fontsize(14) # untuk mengatur huruf
table.scale(1, 4) # berkaitan dengan ukuran table nya (ukuran kolom, ukuran baris)

ax.axis(False) # untuk mengatur axis nya dimunculin apa enggak kalau false tidak kalau

plt.show()
```

### table plot pyplot style

```python
table = plt.table(
    cellText = data_uas, # 
    loc='center' # untuk mengatur lokasi dari table plot
)

table.set_fontsize(14) # untuk mengatur huruf
table.scale(1, 4) # berkaitan dengan ukuran table nya (ukuran kolom, ukuran baris)

ax = plt.gca() # untuk mendapat axes yg sedang aktif (get current axes)
ax.axis(False) # untuk mendesaible/menonaktifkan axes & mengatur axis nya dimunculin apa enggak kalau false tidak kalau

plt.show()
```
