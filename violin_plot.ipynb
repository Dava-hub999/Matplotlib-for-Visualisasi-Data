# Violin Plot / Violin Chart
##### Violin plot adalah jenis plot visualisasi yang menggabungkan elemen dari box plot dan kernel density plot. Ini digunakan untuk menunjukkan distribusi data secara lebih rinci daripada yang dapat diberikan oleh box plot atau histogram biasa.

### Penjelasan Distribusi Normal 
##### distribusi normal mengacu pada bentuk distribusi data yang mengikuti pola yang disebut sebagai distribusi normal atau Gaussian. Distribusi normal adalah jenis distribusi statistik yang memiliki ciri-ciri khusus, seperti memiliki rata-rata yang sama dengan median, serta membentuk kurva simetris yang terpusat di sekitar rata-rata.

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100,  # (loc untuk mengindikasikan nilai mean atau rata2 dari data normal yang dihasilkan)
                        scale=10,  # (scale untuk nilai standard deviasi(semakin tinggi nilai standart deviasi maka semakin luas variasi datanya))
                        size=200 # (banyaknya dataser)
                        )



data
```

### Simple Violin Plot

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

# pengaturan / marker untuk violin plot

```python
plt.violinplot(data,
                showextrema=True, # untuk menampilkan garis tengah serta garis untuk nilai max dan min
                showmeans=False, # untuk memunculkan indikator untuk nilai means
                showmedians=False, # untuk memunculkan indikator untuk nilai medians
                quantiles=[0.25, 0.5, 0.75] # untuk menerima nilaiberupa collection datset atau sekumpulan data
)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

## Horizontal Violin Plot

```python
plt.violinplot(data, vert=False)

plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

# Multiple Violin plot

```python
# Multiple
np.random.seed(10)
data1 = np.random.normal(100, 10, 200) # (nilai rata-rata (mean) dari distribusi normal Artinya: data yang dihasilkan akan memiliki rata-rata 100, sebaran deviasinya yang berarti sebagian besar data akan berada dalam kisaran sekitar 90 hingga 110 (rata-rata  1 deviasi standar), Jumlah datanya)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4] #list untuk penampung keseluruhan 
#labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4'] # list untuk nilai( untuk memberi label di gambar)
```

```python
plt.violinplot(data)

plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```
