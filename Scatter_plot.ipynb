# Scatter Plot / Scatter Chart

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Simple Scatter Plot (contoh sederhana)

```python
# Data x dan y
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x, y, label='Data 1', color='r') # (datax (data yng di plot), datay, berkorelasi dengan legend, untuk color red)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh scatter plot sederhana')

plt.legend()

plt.show()
```

## Multiple scatter plot

```python
# Data x dan y
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x, y, label='Data 1', color='r') # (datax (data yng di plot), datay, berkorelasi dengan legend, untuk color red)
plt.scatter(x2, y2, label='Data 2', color='b') # (datax (data yng di plot), datay, berkorelasi dengan legend, untuk color red)


plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Multiple scatter plot ')

plt.legend()

plt.show()
```

### Pengaturan MArker

```python
# Data x dan y
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(
    x, y, # implemen data
    color='cyan', # warna untuk scatternya
    linewidths=1, # untuk penebalan garis tepi
    marker = 's', # pengaturan bentuk scatter
    edgecolor = 'red', # merubah warna garis tepi
    s = 100 # ukuran/size dari marker atau bentuk nya
)

plt.scatter(
    x2, y2,
    color='yellow',
    linewidths=1,
    marker = '^',
    edgecolor = 'red',
    s = 100
)

plt.show()
```

### Scatter plot Data multidimensi

```python
# untuk import data
from sklearn.datasets import load_iris 
iris = load_iris() # untuk mengembalikan data yg memiliki structur data dict yang berisi keys dan values
features = iris['data'].T
```

```python
features = iris['data'].T # T : digunakan untuk proses transpose(proses membalik kolom dan baris) jadi saling bertukar tempat
features
```

```python
iris['target'] # tujuan data yang akan di scatter
```

```python
iris['target_names'] # pemberian nama pada sebuah target
```

```python
# proses plotting data

plt.scatter(
    features[0],
    features[1],
    alpha =0.2,
    s=100*features[3], # digunakan untuk mengatur size tiap markernya
    c = iris['target'],
    cmap = 'viridis' # untuk mapping atau pemetaan untuk sekumpulan bilangan tertentu 
)

plt.xlabel(iris['feature_names'][0]) # mengambil nama table dari dalam datset pada index 0
plt.ylabel(iris['feature_names'][1])
plt.title('Iris Datset')

plt.colorbar() # untuk keterangan warna

plt.show()

```
