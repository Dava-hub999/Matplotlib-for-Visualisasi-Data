```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# datset sample
n = np.linspace(-3.0, 3.0, 100) # (nilai terkecil 0, nilai terbesar 1, banyak data )

x,y = np.meshgrid(n, n) # menghasilkan data 2 dimensi (data1 jadi x dan data2 jadi y)
z = np.sqrt(x**2 + y**2) # (100 * 100) dan menampung nilai quadratnya

print(x.shape)
print(y.shape)
print(z.shape)
```

# Countor Plot 
### adalah memvisualisasikan data 3 dimensi jadi 2 dimensi

```python
plt.contour(x, y, z)
plt.colorbar() # sebagai metrix untuk membantu menginterpretasikan setiap warna

plt.title('Example Countor Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### Filled Countor Plot (bedanya adalah fullcolor)

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Example Filled Countor Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
