# 3D PLot

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# 3D Axes
fig = plt.figure() # untuk menghasilkan suatu project fitur
ax = plt.axes(projection = '3d') # untuk membuat axes 3D

plt.show()
```

# 3D Line Plot

```python
z = np.linspace(0, 1, 100) # (nilai terkecil 0, nilai terbesar 1, banyak data )
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure() # OO Style
ax = plt.axes(projection = '3d') # menampung fungsi 3D

ax.plot( x,y,z, color='red') # pembentukan line plot

ax.set( # untuk setting seperti biasanya
    title = ' Example 3D Line Plot ',
    xlabel = 'Sumbu X',
    ylabel = 'Sumbu Y ',
    zlabel = 'Sumbu Z'
)

plt.show()
```

# 3D Scatter Plot

```python
z = np.linspace(0, 1, 100) # (nilai terkecil 0, nilai terbesar 1, banyak data )
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure() # OO Style
ax = plt.axes(projection = '3d') # menampung fungsi 3D

ax.scatter( x,y,z, color='green') # pembentukan line plot

ax.set( # untuk setting seperti biasanya
    title = ' Example 3D Scatter Plot ',
    xlabel = 'Sumbu X',
    ylabel = 'Sumbu Y ',
    zlabel = 'Sumbu Z'
)

plt.show()
```
