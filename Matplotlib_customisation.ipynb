# Kustomisasi Matplotlib dengan Style Sheets

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

# Predefined Styles
#### tersedia lengkap disini https://github.com/matplotlib/matplotlib/tree/main/lib/matplotlib/mpl-data/stylelib


```python
plt.style.available
```

```python
# mengambil method dari atas
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 15) # dataset

plt.plot(x, x, 'bs-', label='lin') 
# . Parameter pertama dan kedua adalah x yang sama, sehingga plot akan menghubungkan setiap titik pada sumbu x dengan nilai yang sesuai pada sumbu y (dalam hal ini, nilai x). 

plt.plot(x, np.log(x), 'ro-', label='log') 
#  Parameter pertama adalah x, dan parameter kedua adalah np.log(x) - ini menggambarkan logaritma alami dari setiap nilai dalam array x.

plt.legend()
plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('contoh bar plot')

plt.show()
```

# Menerapkan Temporary Styling / styling sementara

```python
plt.style.use('ggplot') # jenis visualisasi ploting

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'ro-') 
# np.linspace(0, 2 * np.pi): Ini menghasilkan array nilai yang berisi sejumlah titik terdistribusi merata antara 0 dan 2π. Nilai-nilai ini akan digunakan sebagai sumbu x pada plot.

plt.show()
```

```python
# mengindikasikan bahwa styling ini hanya sementara / hanya pada tab ini 
# ada tanda ( with ) nya juga
with plt.style.context('dark_background'): 
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'ro-') 

plt.show()
```
