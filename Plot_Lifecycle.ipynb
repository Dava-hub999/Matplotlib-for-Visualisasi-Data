# Plot Lifecycle (Rangkuman Pola untuk Plooting)

```python
import matplotlib
import os
import matplotlib.pyplot as plt 
import numpy as np
```

# Sample Data

```python
# Struktur Data Dictionary
data={
    'item A' : 1,
    'item B' : 2,
    'item C' : 3,
    'item D' : 4,
    'item E' : 5,
    'item F' : 6,
    'item G' : 7,
    'item H' : 8,
    'item I' : 9,
    'item J' : 10,
}
data
```

```python
# tuple untuk menampung keys
items = tuple(data.keys())
items
```

```python
# tuple untuk menampng values nya
count = tuple(data.values())
count
```

# Simple plot

```python
fig, ax = plt.subplots()

# horizontal bar splot( menampung keys dari datset, menampung values dari datset)
ax.barh(items, count)

plt.show()
```

# Pengaturan Tick Label

```python
fig, ax = plt.subplots()
ax.barh(items, count) # inisialisasi datset

labels = ax.get_xticklabels() # ini digunakan untuk mengakses object dari sumbu X 

# set property untuk mathod get_xticklabels() atau dari variable labels
plt.setp(
    labels,
    rotation=45,
    horizontalalignment='right'
)

plt.show()
```

# Pengaturan Format pada Ticker

#### ticker merujuk pada komponen yang bertanggung jawab untuk mengatur label-label yang ditampilkan pada sumbu grafik (sumbu x dan sumbu y) agar sesuai dengan nilai-nilai data.

```python
from matplotlib.ticker import FuncFormatter
```

```python
# untuk mengonfirmasi nilai ribuan menjadi huruf K
# Example jika 50000 == 50K
def ribuan(x, pos): # (nilai konversi, posisi tickernya)
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

# membuat object formatter dari function
formatter = FuncFormatter(ribuan)
# objek formatter adalah komponen yang digunakan untuk mengontrol tampilan label pada sumbu grafik agar sesuai dengan format yang diinginkan. Objek formatter memungkinkan Anda untuk mengubah cara nilai-nilai data ditampilkan pada sumbu agar lebih informatif dan mudah dibaca.

# di hubungkan agar menjadi sumbu X
ax.xaxis.set_major_formatter(formatter)
# Objek formatter ini kemudian dapat dihubungkan ke sumbu tertentu pada grafik Anda, seperti sumbu x atau sumbu y, untuk mengatur tampilan label-labelnya.

ax.set(
    title = 'Contoh Plot',
    xlabel ='Jumlah',
    ylabel = 'Items'
)

plt.show()
```

### Penambahan Garis (vertical / horizontal Line) pada plot

```python
fig, ax = plt.subplots()
ax.barh(items, count)

## untuk menambah garis vertikal
# ax.axvline(
#     np.mean(count), # bisa diatur berdasarkan nilai x nya 
#     ls='-',
#     color='r'
# )

## untuk menambah garis horizontal
ax.axhline(
    np.mean(count), # bisa diatur berdasarkan nilai x nya 
    ls='-',
    color='r'
)

ax.set(
    title = 'Contoh Plot',
    xlabel ='Jumlah',
    ylabel = 'Items'
)

plt.show
```
