### Visualisasi data yang tersimpan pada pandas data frame(bagian 2)

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

### Dataset

```python
data = {
    'tahun' : [
        '1958', '1963', '1968', '1973', '1978', '1973', '1988', '1993', '1998', '2003', '2008', '2013', '2018'
    ],
    'populasi' : [
        1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000 , 9000, 10000, 11000, 12000, 13000
    ]
}

df = pd.DataFrame(data)

df
```

### LinePlot With pandass

```python
df.plot(kind='line', x='tahun', y='populasi', color='red') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('Line Plot')
plt.xlabel('Populasi') 
plt.ylabel('Tahun')

plt.show()
```

# Bar Plot With pandas

```python
# Sample Dataset

data = {
    'negara' : [
        'US', 'Singapore', 'Germany', 'UK', 'Japan'
    ],
    'GDP' : [
        62606, 100345, 52559, 45705, 44227
    ]
}

df = pd.DataFrame(data)

df
```

```python
df.plot(kind='bar', x='negara', y='GDP', color='red') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('vertikal Bar Plot ')
plt.xlabel('Negara') 
plt.ylabel('GDP dalam USD')

plt.show()
```

```python
df.plot(kind='barh', x='negara', y='GDP', color='red') # (menentukan jenis plotting, data X, data Y, warna)

plt.title('vertikal Bar Plot ')
plt.xlabel('Negara') 
plt.ylabel('GDP dalam USD')

plt.show()
```

```python
df.plot(kind='barh', x='negara', y='GDP', color='red', legend=False) # (menentukan jenis plotting, data X, data Y, warna)

plt.title('vertikal Bar Plot ')
plt.xlabel('Negara') 
plt.ylabel('GDP dalam USD')

plt.show()
```

## Scatter Plot with pandas

```python
# Sample Dataset

data = {
    'negara' : [
        'US', 'Singapore', 'Germany', 'UK', 'Japan'
    ],
    'GDP' : [
        62606, 100345, 52559, 45705, 44227
    ],
    'Life_Expectancy' :[
       79.24, 82.84, 80.84, 81.40, 83.62 
    ]
}

df = pd.DataFrame(data)

df
```

```python
df.plot(kind='scatter', x='GDP', y='Life_Expectancy', color='red') # (menentukan jenis plotting, data X, data Y, warna)

plt.title(' GDP dan Life_Expectancy ')
plt.xlabel('GDP dan USD') 
plt.ylabel('Life_Expectancy')

plt.show()
```

# Pie Plot with Pandas

```python
data = {
    'benua' : [
        'South America', 'Oceania', 'North America', 'Europe', 'Asia', 'Antartica', 'Africa'
    ],
    'populasi' : [
        123751, 2732740, 347192, 4138249, 5382749, 63279849, 7947481
    ]
}

df = pd.DataFrame(data)

df
```

```python
df.plot(kind='pie', y='populasi', figsize=(6,6)) # (menentukan jenis plotting, data y, terkait ukuran figure)

plt.title('Populasi di tiap benua')
# plt.xlabel('Negara') 
# plt.ylabel('GDP dalam USD')

plt.show()
```

```python
df = df.set_index('benua')
df
```

```python
df.plot(kind='pie', y='populasi', figsize=(6,6)) # (menentukan jenis plotting, data y, terkait ukuran figure)

plt.title('Populasi di tiap benua')
# plt.xlabel('Negara') 
# plt.ylabel('GDP dalam USD')

plt.show()
```

## Box Plot with pandas

```python
data = {
    'benua' : [
        'South America', 'Oceania', 'North America', 'Europe', 'Asia', 'Antartica', 'Africa'
    ],
    'populasi' : [
        123751, 2732740, 347192, 4138249, 5382749, 63279849, 7947481
    ]
}

df = pd.DataFrame(data)

df
```
