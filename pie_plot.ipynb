# Pie Plot atau Pie Chart

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Simple pie plot / sederhana

```python
kategori = ['A', 'B', 'C', 'D', 'E'] # kategori dan data akan saling berasosiasi
data = [215, 130, 245, 210, 225]
```

```python
plt.pie( # memanggil fungsi yang membutuhkan beberapa nilai parameter
    data, # manggil data untuk int
    labels=kategori, # berasosiasi dengan kategori (bersifat opsional apabila di beri variabel labels)
    autopct='%1.1f%%', # digunakan untuk menampilkan persentasi untuk setiap area (opsional)
    startangle= 90 # untuk mengatur rotasi (opsional)
    )

plt.title(' Contoh Pie Plot ')

plt.show()
```

# Pengaturan Legend dan Color


```python
kategori = ['A', 'B', 'C', 'D', 'E'] # kategori dan data akan saling berasosiasi
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] # menampung list untuk merepresentasikan warna dari tiap kategori

# untuk 3 data ini saling berasosiasi
```

```python
plt.pie( # memanggil fungsi yang membutuhkan beberapa nilai parameter
    data, # manggil data untuk int
    labels=kategori, # berasosiasi dengan kategori
    colors = warna, # berasosiasi dengan colors
    autopct='%1.1f%%', # digunakan untuk menampilkan persentasi untuk setiap area (opsional)
    startangle= 60 # untuk mengatur rotasi (opsional)
    )

plt.legend()

plt.title(' Contoh Pie Plot untuk mengatur legend ')

plt.show()
```

## Exploded pie plot 
### fungsinya untuk memisahkan pie plot


```python
kategori = ['A', 'B', 'C', 'D', 'E'] # kategori dan data akan saling berasosiasi
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] # menampung list untuk merepresentasikan warna dari tiap kategori
explode_var = [0., 0., 0.2, 0., 0.08] # yang akan berkorelasi dengan exploded pie plot
# explode var untuk tiap parameter memiliki nilai masing masing dan yang tidak akan di dipisahkan diberi nilai (0.)
# makin besar nilai explode var nya maka makin jauh juga jaraknya

# untuk 4 data ini saling berasosiasi
```

```python
plt.pie( # memanggil fungsi yang membutuhkan beberapa nilai parameter
    data, # manggil data untuk int
    labels=kategori, # berasosiasi dengan kategori
    colors = warna, # berasosiasi dengan colors
    explode = explode_var, # digunakan untuk memisah antar pie plot (berasosiasi dengan 3 data lain)
    autopct='%1.1f%%', # digunakan untuk menampilkan persentasi untuk setiap area (opsional)
    startangle= 60 # untuk mengatur rotasi (opsional)
    )

plt.legend()

plt.title(' Contoh Pie Plot untuk mengatur exploded pie plot ')

plt.show()
```
