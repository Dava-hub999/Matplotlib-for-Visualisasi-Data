# Twin Axes


```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Study kasus Twin Axes 1 

```python
x = np.linspace(1, 10, 25) # (start, stop, banyaknya atau jumlahnya) jadi randoman
x
```

```python
fig, ax1 = plt.subplots() # OO Style

ax1.plot(x, np.exp(x), 'bs-', label='exp') # nilai exp / Nilai ekspektasi adalah konsep dalam statistik dan teori peluang yang mengukur nilai rata-rata yang diharapkan dari suatu peristiwa atau variabel acak.
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx() # pemanggilan twin x
ax2.plot(x, np.log(x), 'ro-', label='log') #  logaritma didefinisikan sebagai pangkat yang diperlukan untuk menghasilkan suatu bilangan tertentu dari bilangan lain, yang disebut basis, dalam suatu operasi eksponensial.
ax2.set_ylabel('log')

fig.suptitle('contoh twin access') # memberi judul
fig.legend(loc='upper left') # mangelola tata letak legend nya

plt.show()
```

# Study kasus Twin Axes part 2


```python
x = np.random.randint(16, 40, size=30) # (menentukan nilai terkecil, menentukan nilai terbesar, jumlah datanya) dalam lingkup int
x
```

```python
def C2F(celcius=0): # fungsi untuk menerima nilai dalam drajat celcius konversi ke fahrenheit
    return (celcius*1.8) + 32  # hasil di return

def konversi_sumbu(ax1): # konversi matrix dari sumbu x oertama ke yg kdua
    y1, y2 = ax1.get_ylim() # Baris ini mengambil batas sumbu y (nilai minimum dan maksimum) dari sumbu "ax1" yang diberikan, dan menyimpannya dalam variabel "y1" dan "y2".
    ax2.set_ylim(C2F(y1), C2F(y2)) # C2F(y1) mengonversi batas sumbu y y1 dari Celsius ke Fahrenheit dan menggunakannya sebagai batas minimum pada sumbu y untuk plot pada "ax2". Hal yang sama juga berlaku untuk C2F(y2) yang digunakan sebagai batas maksimum.
    ax2.figure.canvas.draw() #  Baris ini memaksa gambar (plot) yang terkait dengan "ax2" untuk diperbarui sehingga perubahan batas sumbu y baru akan terlihat dalam plot.

#  ylim digunakan untuk mengatur batas sumbu pada plot, dan dapat digunakan bersama dengan callback untuk mengendalikan tampilan plot berdasarkan perubahan tertentu.

fig, ax1 = plt.subplots() # OO Style
ax2 = ax1.twinx() # twin axes dari sumbu ax1 ke ax2

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
# ax1.callbacks.connect('ylim_changed', konversi_sumbu), Anda sedang menghubungkan fungsi konversi_sumbu dengan peristiwa perubahan batas sumbu y (ylim_changed) pada objek sumbu ax1. Jadi, setiap kali batas sumbu y pada sumbu ax1 berubah, fungsi konversi_sumbu akan dieksekusi. Dengan cara ini, Anda dapat mengendalikan tindakan yang dilakukan saat batas sumbu y berubah pada plot tersebut.
ax1.plot(x) # data dari x

ax1.set_xlabel('Hari keberapa')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('fahrenheit')

fig.suptitle('Temperatur Udara')
```
