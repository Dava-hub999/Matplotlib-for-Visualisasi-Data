# Membuat barplot(diagram batang)

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
```

## Simple Bar Plot

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data) # parameter 1 (kategori) biasanya kategorical, 
# paremeter 2 (data) sederet nilai yang berasosiasi dengan setiap item yang ada di parameter perta,a
# pemanggilan data dan methode bar plot

plt.grid()

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title(' Contoh Bar Plot ')

plt.show()
```

##### perbedaan bar plot dan histogram
##### histogram digunakan untuk plooting data yang bertype numerik dan bersifat continus
##### bar plot digunakan untuk plooting data yang bertype kategorical atau discrub 

## Pengaturan Grid dan Color

```python
plt.bar(kategori,
        data,
        color='y',
        alpha= 0.75
        ) 

plt.grid(linestyle='--', # mengubah menjadi garis2
        linewidth=2, # mengatur ketebalan
        axis = 'y', # pilihan x, y, both
        alpha= 0.50
        )

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title(' Contoh Bar Plot dan penerapan grid and color ')

plt.show()
```

```python
plt.barh(kategori,
        data
        )
#plt barh(kategori[::-1], data(::-1))

plt.grid(linestyle='--', # mengubah menjadi garis2
        linewidth=2, # mengatur ketebalan
        axis = 'y', # pilihan x, y, both
        alpha= 0.50
        )

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title(' Contoh Horisontal Bar Plot ')

plt.show()
```

## Stacker bar plot

#### digunakan untuk melakukan ploting 2 datset berbeda tapi dengan 1 kategori yang sama

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,
        data1,
        label='data1'
        )

plt.bar(kategori,
        data2,
        label='data2',
        bottom=data1 # untuk menentukan letak dari inisialisasi dataset yang ke 2
        )

plt.legend()

plt.grid(linestyle='--', # mengubah menjadi garis2
        linewidth=2, # mengatur ketebalan
        axis = 'y', # pilihan x, y, both
        alpha= 0.75 # transparansi
        )

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title(' Contoh Stacked Bar Plot ')

plt.show()
```

# Grouped bar plot

### barplot dari 2 datset berbeda akan dikelompokkan berdasarkan kategori dan akan di plooting berdampingan satu dengan yang lain

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori)) # digunkan untuk menampung suatu array dari(np.arange) nah fungsinya untuk menghasilkan suatu array yang punya suatu deret bilangan darinilai yang ada dalam (len(kategori)) / jumlah index dari kategroi
x
```

```python
width = 0.35 # untuk mengatur jarak antar barplot

plt.bar(x-width/2, data1, width, label='Data 1') 
plt.bar(x+width/2, data2, width, label='Data 2') 
# berasosiasi dengan nilai x dari(len(kategori)) sekaligus mengatur width nya agar tidak tumpuk jadi ada +(data2) dan -(data1) atau data1(kiri) data2(kanan) 

plt.xticks(x, kategori) # digunakan untuk mengatur menginisiasi labelnya(penampung index, diubah ke dalam kategori) 

plt.legend()

plt.grid(linestyle='--', # mengubah menjadi garis2
        linewidth=1, # mengatur ketebalan
        axis = 'y', # pilihan x, y, both
        alpha= 0.75 # transparansi
        )

plt.xlabel('kategori')
plt.ylabel('jumlah')
plt.title(' Contoh Grouped Bar Plot ')

plt.show()
```
