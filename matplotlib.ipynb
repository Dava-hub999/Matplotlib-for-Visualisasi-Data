## Pengertian Dasar
### matplotlib merupakan visualisasi data dalam bentuk diagram yang terdapat komponen inti yaitu (figure, axes, axis)
#### 1. figure : merupakan container besar jadi hanya kayak lembar kosong dan bisa terdiri dari beberapa axes
#### 2. axes : area gambar yang ada dalam figure(area plooting) atau area diagram nya
#### 3. axis : pengindikasian sumbu (x, y atau lebih) jadi bisa dikatakan isi dari axes

```python
# import module
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# contoh ploting sederhana

# contoh data simple
x = [1,2,3,4]
y = [1,4,2,3]

fig, ax = plt.subplots() # membuat figure dan axes
ax.plot(x, y) # membuat axis / plotting data pada axes
```

```python
# contoh hanya menggunakan plot tanpa penginisialisasian fig dan ax nya

# contoh data simple
x = [1,2,3,4]
y = [1,4,2,3]

plt.plot(x, y)
```

### contoh pembuatan figure dan axes

```python
fig = plt.figure() # membuat figure tanpa axes
```

```python
fig, ax = plt.subplots() # pembuatan figure dan axes tanpa axis(pengidentifikasian sumbu)
```

```python
fig, ax = plt.subplots(2, 3) # figure dengan 2x3 grid axes (belum ada axisnya ini)
```

# 2 cara penggunaan matplotlib

### 1. OO Style ( object oriented Style)

```python
x = np.linspace(0, 2, 100) # (Start, Stop, Step)
x
```

```python
fig , ax = plt.subplots() # pembuatan figure dan axes

# plotting 3 variant data pada axes
ax.plot (x, x, label='linear', color='pink')
ax.plot (x, x**2, label='quadratic', color='green')
ax.plot (x, x, label='cubic', color='black')

ax.set_xlabel ('x label') # menyertakan x label pada axes
ax.set_ylabel('y label') # menyertakan y label pada axes
ax.set_title('Simple plot') # menyertakan tit;e label pada axes
ax.legend() # menambahkan legend atau keterangan
```

### 2. pyplot style

```python
fig , ax = plt.subplots() # pembuatan figure dan axes

# plotting 3 variant data pada axes
plt.plot (x, x, label='linear', color='pink')
plt.plot (x, x**2, label='quadratic', color='green')
plt.plot (x, x, label='cubic', color='black')

plt.xlabel ('x label') # menyertakan x label pada axes
plt.ylabel('y label') # menyertakan y label pada axes
plt.title('Simple plot') # menyertakan tit;e label pada axes
plt.legend() # menambahkan legend atau keterangan
```

### Perbedaan antara keduanya adalah kalau oos(object oriented style) dia harus inisialisasi fig, ax dulu dan menggunakan kata kunci ax kalau pyplot menggunkana plt
